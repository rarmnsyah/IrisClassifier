```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
target = df['Species']
target_names = target.unique()
sample = df.drop('Species', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(sample, target, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def kNeighborsClassifier(X_train, y_train):
    model = KNeighborsClassifier(n_neighbors = 3)
    model.fit(X_train, y_train)
    return model
```

```python
from sklearn.ensemble import RandomForestClassifier

def randomForests(X_train, y_train, n):
    model = RandomForestClassifier(n_estimators = n)
    model.fit(X_train, y_train)
    return model
```

```python
from sklearn.linear_model import LogisticRegression

def logisticRegression(X_train, y_train):
    model = RandomForestClassifier(random_state = 0)
    model.fit(X_train, y_train)
    return model
```

```python
model = kNeighborsClassifier(X_train, y_train)
pred = model.predict(X_test)

print(sum(y_test == pred)/ len(pred))
```

```python
model = randomForests(X_train, y_train, 100)
pred = model.predict(X_test)

print(sum(y_test == pred)/ len(pred))
```

```python
model = logisticRegression(X_train, y_train)
pred = model.predict(X_test)

print(sum(y_test == pred)/ len(pred))
```
